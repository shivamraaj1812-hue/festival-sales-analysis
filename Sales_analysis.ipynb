```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt #visualizing data
%matplotlib inline
import seaborn as sns
```

```python
df=pd.read_csv('diwali_sales.csv',encoding='unicode escape')
df.shape
```

```python
df.head()
```

```python
df.info()
```

```python
df.drop(['Status','unnamed1'],axis=1,inplace=True)
```

```python
df.shape[1]
```

```python
pd.isnull(df).sum()
```

```python
df.dropna(inplace=True)
```

```python
df['Amount']=df['Amount'].astype('int')
#changing data type for better analysis
```

```python
#Gender anlaysis
df.columns
```

```python
ax=sns.countplot(x='Gender',data=df)

for bars in ax.containers:
    ax.bar_label(bars)


plt.show()
#This graph analysis shows that female purchased more than male
```

```python
sales_gen=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
sns.barplot(x='Gender',y='Amount',data=sales_gen)
plt.show()
```

```python
#From above graph we can see that most of the buyers are females and even their purchasing power is more
```
